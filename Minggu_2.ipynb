```python
import pandas as pd  # alias umum untuk DataFrame dan I/O
import seaborn as sns  # dataset contoh + visualisasi

df_tips = sns.load_dataset("tips")  # DataFrame restoran (ujung proporsi / regresi)
df_titanic = sns.load_dataset("titanic")  # DataFrame penumpang (klasifikasi umum)

# Aktifkan baris di bawah untuk ringkas dimensi dan tipe tips:
print("tips", df_tips.shape)
print(df_tips.dtypes.head())
print("10 data dari tips dataset : ")
df_tips.head(10)  # tampilkan tips jika perlu
# df_titanic.head(10)  # output terakhir: tampilan HTML 10 baris pertama Titanic
```

```python
print(df_titanic.info())  # ringkasan kolom, non-null count, dtype
# df_titanic.describe(include="all").T.head(14)  # statistik deskriptif (aktifkan untuk eksplorasi)
```

```python
from sklearn.datasets import fetch_openml  # akses dataset publik OpenML

iris_oml = fetch_openml(name="iris", version=1, as_frame=True)  # as_frame=True -> DataFrame di .frame
df_iris = iris_oml.frame  # fitur + target dalam satu tabel
print(df_iris.shape, df_iris.columns.tolist())  # dimensi + nama kolom
df_iris.head()  # lima baris pertama
```

```python
def ringkas(nama, frame):  # nama: label; frame: DataFrame
    return {
        "nama": nama,
        "baris": len(frame),  # jumlah baris
        "kolom": frame.shape[1],  # jumlah kolom
        "missing_total": int(frame.isna().sum().sum()),  # total sel NaN
    }


print(ringkas("tips", df_tips))
print(ringkas("titanic", df_titanic))
print(ringkas("iris_oml", df_iris))
```
